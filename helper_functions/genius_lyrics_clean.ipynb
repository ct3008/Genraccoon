```python
import pandas as pd
import re
import string

df = pd.read_csv("lyrics_found.csv")
```

```python
def clean_lyrics(lyrics):
    if pd.isna(lyrics):
        return ""
    
    # Convert to string and remove line breaks
    text = str(lyrics).replace("\n", " ")
    
    # Remove everything before the first " Lyrics"
    text = re.sub(r'^.*?Lyrics', '', text, flags=re.IGNORECASE
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = " ".join(text.split())
    
    return text

df['lyrics'] = df['genius_lyrics'].apply(clean_lyrics)
```

```python
df.drop(columns=['genius_lyrics'], inplace=True)
```

```python
merge_map = {
    "Rock": ["Rock", "Pop Rock", "Rock Metal", "Rock Pop", "Rock Folk", "Rock Electronic", "Punk Rock"],
    "Pop": ["Pop", "Pop_Rock", "Pop_Rock Folk", "Pop Electronic"],
    "Rap": ["Rap"],
    "Electronic": ["Electronic", "Electronic Pop", "Electronic Rock"],
    "Jazz": ["Jazz"],
    "Metal": ["Metal", "Rock Metal"],
    "Blues": ["Blues", "Rock Blues"],
    "Punk": ["Punk", "Rock Punk"],
    "RnB": ["RnB"],
    "Country": ["Country"],
    "Reggae": ["Reggae"],
    "Latin": ["Latin"],
    "Folk": ["Folk"],
    "World": ["World"],
    "New Age": ["New Age"],
}

# Create reverse mapping
flat_map = {g: new for new, old_list in merge_map.items() for g in old_list}
df['genre'] = df['genre'].map(flat_map).fillna(df['genre'])  # fallback to original if unmapped
```
